## Merging the datasets

```python
movies=movies.merge(credits,on='title')
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

## Handling missing and duplicate values

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
movies.rename(columns={'genres':'genre'},inplace=True)
```

```python
movies.info()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genre']=movies['genre'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.rename(columns={'genre':'genres'},inplace=True)
```

```python
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast'][0]
```

```python
movies.head()
```

```python
movies.iloc[0].crew
```

```python
def fetch_director(text):
    L = []
    
    for i in ast.literal_eval(text):
        if i['job']=='Director':
        
            L.append(i['name'])
        
    return L 
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Sam Worthington-----> SamWorthington
```

```python
def combine(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
        
```

```python
movies['genres']=movies['genres'].apply(combine)
movies['keywords']=movies['keywords'].apply(combine)
movies['cast']=movies['cast'].apply(combine)
movies['crew']=movies['crew'].apply(combine)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new=movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags']=new['tags'].apply(lambda x: " ".join(x))
```

```python
new.head()
```

```python
new['tags']=new['tags'].apply(lambda x:x.lower())
```

```python
new['tags'][0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
new.head()
```

# Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new['tags'])
```

```python
np.array(vector)
```

```python
vector
```

```python
vector.shape
```

```python
np.set_printoptions(threshold=np.inf)
```

# Calculating Cosine_Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity[1]
```

```python
similarity[1].shape
```

```python
new
```

```python
new[new['title']=='Avatar'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('The Dark Knight Rises')
```
